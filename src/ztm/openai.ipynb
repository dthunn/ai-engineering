```python
from dotenv import load_dotenv, find_dotenv
from openai import OpenAI

load_dotenv(find_dotenv(), override=True)
```

```python
client = OpenAI()
```

```python
response = client.chat.completions.create(
    model='gpt-5.2',
    messages= [
        {'role': 'system', 'content': 'You are a helpful assistant.'},
        {'role': 'user', 'content': 'Write a short article about coffee.'}
    ],
    temperature=0.8,  # Controls randomness: 0=deterministic, 2=very creative
    # seed=1234,  # Fix seed for reproducible outputs (same seed → same result)
    # max_completion_tokens=1000,  # Hard cap on the number of tokens in the reply
    # n=2,  # Generate multiple independent completions
#     stop=[';', '.', '\n']
    # frequency_penalty=0,  # Penalizes repeated tokens; range [-2, 2], default 0
    # presence_penalty=0    # Encourages new topics; range [-2, 2], default 0
)

# Print the text content of the first completion choice.
print(response.choices[0].message.content)
```
